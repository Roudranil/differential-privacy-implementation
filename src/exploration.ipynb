# Smartnoise


```python
import snsql
from snsql import Privacy
import pandas as pd
```

```python
privacy = Privacy(epsilon=0.5, delta=0.001)
data_path = "../data/broadband_data_zipcode.csv"
metadata_path = "../data/broadband_data_zipcode_metadata.yaml"
zipcode_data = pd.read_csv(data_path)
reader = snsql.from_df(zipcode_data, privacy=privacy, metadata=metadata_path)
```

```python
result = reader.execute(
    "SELECT COUNTY_ID, AVG(BROADBAND_USAGE) as BROADBAND_USAGE from Broadband.Zipcodes GROUP BY COUNTY_ID"
)
```

```python
zipcode_data.groupby("COUNTY_ID")["BROADBAND_USAGE"].mean()
```
